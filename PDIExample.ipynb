```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Ruta de la imagen
image_path = "/content/cat.jpg"
```

```python
# Leer la imagen
image = cv2.imread(image_path, 0)
print(image)
```

```python
# Verificar si la imagen se ley√≥ correctamente
if image is None:
    print("No se pudo cargar la imagen")
else:
    print(f"Imagen cargada exitosamente. Dimensiones: {image.shape}")
```

```python
# Mostrar la imagen en una ventana
cv2_imshow(image)

# Esperar a que el usuario cierre la ventana
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
# Ruta para guardar la nueva imagen
output_path = "/content/salida.jpg"

# Guardar la imagen
success = cv2.imwrite(output_path, image)

if success:
    print(f"Imagen guardada exitosamente en {output_path}")
else:
    print("No se pudo guardar la imagen")
```
